```python
import plotdata
import matplotlib.pyplot as plt
```

```python
df = plotdata.metric_vs_property(
    agent='Arty',
    scenario='MarketSquare',
    scenario_parameter=None,
    property='agent_parameter',
    metrics=['completion_time', 'max_iteration_time']
)
def plot_two(df):
    "Plots two columns against each other using secondary y axis"
    assert(len(df.columns) == 2)
    
    plt.figure()
    df.iloc[:, 0].plot(label="foo")
    df.iloc[:, 1].plot(secondary_y=True, label="bar")
    
plot_two(df)
```
