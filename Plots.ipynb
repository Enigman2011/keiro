##Method

There are essentially two different variables we care about (in order of importance):
* Completion time (in simulated time)
* Iteration time (in real time)

It's assumed that the only admissable results are those without any collisions. Paths with collisions will still be notedIn the case that there are a few, hard-to-avoid collisions.
Completion time is chosen as the important metric as it is implementation and platform independent. Iteration time is subject to many kinds of low-level optimization and will depend on things like implementation language, hardware specification, unrelated CPU memory load etc.

It is however important that algorithms still operate in the realm of feasibility. While a (near-)complete search of the problem space could potentially yield close to optimal results, such a search is not feasible to do using small commodity hardware, no matter the number of low-level optimizations performed.

### Generic helper functions

```python
METRICS = ["completion_time", "avg_iteration_time",
           "collisions", "max_iteration_time"]

def get_data(agent, scenario, scenario_parameter, ids=None):
    if ids:
        q = partial(
            Record.objects.filter,
            id__in=ids
        )
    else:
        q = Record.objects.filter

    qs = q(
        agent=agent,
        scenario=scenario,
        scenario_parameter=scenario_parameter
    )
    return pd.DataFrame(list(qs.values()))


```

```python
def plot_two(df):
    "Plots two columns against each other using secondary y axis"
    assert(len(df.columns) == 2)
    
    plt.figure(figsize=(14, 5))
    df.iloc[:, 0].plot(label="foo")
    df.iloc[:, 1].plot(secondary_y=True, label=df.columns[1])
    plt.legend()
```

```python
def stat_cols(df, metric_names, groupby="agent_parameter"):
    """Creates a new dataframe with average, max and min of columns
    provided in metric_names"""
    groups = df.groupby(groupby)
    stats = {}
    for metric in metric_names:
        stats["avg_" + metric] = groups[metric].aggregate(np.average)
        stats["max_" + metric] = groups[metric].aggregate(max)
        stats["min_" + metric] = groups[metric].aggregate(max)
    stats_df = pd.DataFrame(stats)
    return stats_df

arty_marketsquare = get_data("Arty", "MarketSquare", None)
arty_marketsquare_agent_parameter_agg = stat_cols(arty_marketsquare, METRICS, "agent_parameter")
plot_two(arty_marketsquare_agent_parameter_agg[["avg_completion_time", "max_max_iteration_time"]] )
```
