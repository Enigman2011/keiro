##Method

There are essentially two different variables we care about (in order of importance):
* Completion time (in simulated time, or iterations)
* Iteration time (real time it takes to compute one iteration)

It's assumed that the only admissable results are those without any collisions. Paths with collisions will still be noted. In the case that there are a few, hard-to-avoid collisions.
Completion time is chosen as the important metric as it is implementation and platform independent. Iteration time is subject to many kinds of low-level optimization and will depend on things like implementation language, hardware specification, unrelated CPU memory load etc.

It is however important that algorithms still operate in the realm of feasibility. While a (near-)complete search of the problem space could potentially yield close to optimal results, such a search is not feasible to do using small commodity hardware, no matter the number of low-level optimizations performed.

### Generic helper functions

```python
METRICS = ["completion_time", "avg_iteration_time",
           "collisions", "max_iteration_time"]

def get_data(**query_args):
    qs = Record.objects.filter(**query_args)
    return pd.DataFrame(list(qs.values()))

def data_is_valid(data):
    "Returns True if only one row per (agent parameter, seed) combination"
    group_sizes = data.groupby(["agent_parameter", "seed"]).size()
    valid = (group_sizes == 1).all()
    if not valid:
        print "Errors in:"
        print group_sizes[group_sizes != 1]
    return valid

def stat_cols(df, metric_names, groupby="agent_parameter"):
    """Creates a new dataframe with average, max and min of columns
    provided in metric_names"""
    groups = df.groupby(groupby)
    stats = {}
    for metric in metric_names:
        stats["avg_" + metric] = groups[metric].aggregate(np.average)
        stats["max_" + metric] = groups[metric].aggregate(max)
        stats["min_" + metric] = groups[metric].aggregate(min)
    stats_df = pd.DataFrame(stats)
    return stats_df
```

### ART

Since the performance of ART may depend to a large extent on the algorithm parameter used (specifying the size of the global search graph), characteristics of ART in various scenarios as a function of the algorithm parameter are examined.

```python
def plot_min_max_avg(dataframe, groupby, column, title=None):
    groups = dataframe.groupby(groupby)
    
    mins = groups[column].aggregate(min)
    maxs = groups[column].aggregate(max)
    avgs = groups[column].aggregate(np.average)

    pyplot.title("Foo")  # no effect?
    pyplot.ylabel(column)
    pyplot.xlabel(groupby)
    pyplot.errorbar(
        avgs.index, avgs, marker='o',
        yerr=np.row_stack((avgs - mins, maxs - avgs))
    )
    if title is not None:
        pyplot.title(title)
        
        
def agent_parameter_vs_performance(agent, scenario, scenario_parameter=None):
    fig = pyplot.figure(figsize=(16, 10))
    df = get_data(
        agent=agent,
        scenario=scenario,
        scenario_parameter=scenario_parameter
    )
    assert data_is_valid(df)
    pyplot.subplot(211)
    if scenario_parameter is not None:
        parameter_string = "(p={})".format(scenario_parameter)
    else:
        parameter_string = ""
    title1="{scenario}{parameter_string}: {agent} completion time (average, min, max)".format(
        scenario=scenario,
        parameter_string=parameter_string,
        agent=agent
    )
    title2="{scenario}{parameter_string}: {agent} max iteration time (average, min, max)".format(
        scenario=scenario,
        parameter_string=parameter_string,
        agent=agent
    )
    plot_min_max_avg(df, "agent_parameter", "completion_time", title1)
    pyplot.subplot(212)
    plot_min_max_avg(df, "agent_parameter", "max_iteration_time", title2)
    pyplot.savefig('pyplots/{}_{}{}.pdf'.format(agent, scenario, parameter_string))
 
```
